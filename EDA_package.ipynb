```python
import numpy as np
import pandas as pd
import zipfile
import wget
from dataprep.eda import plot, plot_correlation, plot_missing
```

```python
# downloading the data directly from url
data_url = "http://archive.ics.uci.edu/ml/machine-learning-databases/00462/drugsCom_raw.zip"
wget.download(data_url,'drugsCom_raw.zip')
# unzipping the contents
zf = zipfile.ZipFile('drugsCom_raw.zip')
# checking the contents
zipfile.ZipFile.namelist(zf)
```

```python
# reading both the train and test data
train_raw = pd.read_csv( zf.open(zipfile.ZipFile.namelist(zf)[1]) , sep = "\t" )
test_raw = pd.read_csv( zf.open(zipfile.ZipFile.namelist(zf)[0]) , sep = "\t" )
train_raw.head()
```

```python
# checking the content of train & test data
print ( 'Train Columns: ' , train_raw.columns ,'\nTest Columns: ', test_raw.columns )
print ( 'No of unique values with unnamed variable:', train_raw['Unnamed: 0'].nunique() )
print ( 'Unique values with rating variable:', train_raw['rating'].nunique() )
```

# Basic Package:  dataprep.eda :  plot
## To analyse basic characteristics of the dataset like distribution plot, bar chart etc.

```python
plot(train_raw)
# Check out the information of missing against 'condition' variable 
```

```python
plot(train_raw, 'usefulCount'  )
# We also have KDE, QQ normal & Box plot.
```

```python
train_raw['usefulCount1'] = train_raw['usefulCount'].clip(0,300)
plot(train_raw, 'usefulCount1', bins = 20 )
# bins can be changed from default 10 to required.
```

```python
plot(train_raw,'condition')
# pie chart is an effective way to understand the wholestic picture
```

```python
plot(train_raw,'rating','usefulCount1')
# Apart from scatter plot, Box plot and hexbins are important to peform bivariate analysis
```

# Correlation Package:  dataprep.eda :  plot_correlation
## To analyse correlation among the variables of the dataset

```python
plot_correlation(train_raw)
# Calculates Spearman Correlations too. Heat map helps to interpret results quickly.
# plot_correlation(train_raw, k = 10); top 10 attribute correlation in the dataset
# plot_correlation(train_raw, 'usefulCount', value_range = [-1,.3]); tweak the ploting range
```

```python
plot_correlation(train_raw, x = 'usefulCount', y = 'rating' , k = 10 )
# Also highlights the influential points in getting the correlation. Here k = 10 for 10 such points.
```

```python
# Missing Package:  dataprep.eda :  plot_missing
## To analyse missing data points and the impact post removing.
```

```python
plot_missing(train_raw)
# notice the ratio adjacent to the variable names
```

```python
plot_missing(train_raw, "condition","rating")
# count of rows with and without dropping the missing values
# No visible change in box plot between original and dropped missing variables.
# Apart from Histogram we have PDF & CDF too. 
```
